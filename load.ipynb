```python
import pandas as pd
from sqlalchemy import inspect, create_engine
```

```python
HOSTNAME = "localhost"
PORT = 5432
DATABASE = "top_charts3"
USERNAME = "postgres"
PASSWORD = "password"

path = f'postgresql+psycopg2://{USERNAME}:{PASSWORD}@{HOSTNAME}:{PORT}/{DATABASE}'
engine = create_engine(path)
```

```python
inspector = inspect(engine)

for table in inspector.get_table_names():
    columns = inspector.get_columns(table)
    print(table)
    print()
    for column in columns:
        print(column["name"], column["type"])
    
    # or put it as a dataframe
    display(pd.DataFrame(columns))
    
    print()
```

```python
#load artist
artist_csv = pd.read_csv("final_csvs/artists.csv") 

artist_csv.to_sql(name='artist', con=engine, if_exists='append', index=False) 
```

```python
#extract data
artist_df= pd.read_sql("select * from artist;", con=engine)
artist_df.head()
```
