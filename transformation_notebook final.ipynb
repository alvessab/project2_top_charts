```python
import pandas as pd
import numpy as np
```

```python
#read in spotify data
path = "Spotify_data.csv"

spotify = pd.read_csv(path)
spotify.head()
```

```python
#read in apple data
path = "apple_data.csv"

apple = pd.read_csv(path)
apple.head()
```

```python
#join together csv files
main = pd.merge(spotify, apple, on="song_name", how="outer")
main.head()
```

```python
#fill in artist column from two original columns
#https://stackoverflow.com/questions/55117802/how-can-i-check-if-a-value-is-nan-or-a-list
for i, row in main.iterrows():
    value = row.artist_y
    if pd.isnull(value):
        main["artist"]= main.artist_x
main.head()
```

```python
main.info()
```

```python
#review rows that have no artist
mask = main.artist.isnull()
main.loc[mask]
```

# artist table

```python
spotify_artist = list(spotify.artist)
```

```python
apple_artist = list(apple.artist)
```

```python
artists_org = spotify_artist + apple_artist
artists_org
```

```python
len(artists_org)
```

```python
len(np.unique(artists_org))
```

```python
artist_list = list(np.unique(artists_org))
artist_list
```

```python
artists = pd.DataFrame()
artists["artist_name"] = artist_list
artists.head()
```

```python
artists["artist_ID"] = artists.index +1
artists.head()
```

```python
#create csv
artists.to_csv("csvs/artists.csv")
```

```python
#combine artist columns in main
main= pd.merge(main, artists, left_on="artist", right_on="artist_name", how="left")
main.head()
```

```python
main.info()
```

# Songs

```python
spotify_songs = list(spotify.song_name)
apple_songs = list(apple.song_name)
songs_org = spotify_songs + apple_songs
songs_org
```

```python
len(songs_org)
```

```python
len(np.unique(songs_org))
```

```python
songs_list = list(np.unique(songs_org))
```

```python
songs = pd.DataFrame()
songs["song_name"] = songs_list
songs.head()
```

```python
songs["song_ID"] = songs.index +1
songs.head()
```

```python
len(songs)
```

```python
#merge ID onto main 
main = pd.merge(main, songs, on="song_name", how="left")
main.head()
```

```python
main.info()
```

```python
#retain only song specific information from main joined table
main_songs = main[["song_name", "artist_ID", "album_ID", "song_length"]]
main_songs.head()
```

```python
#merge into songs df
songs = pd.merge(songs, main_songs, on="song_name")
songs.head()
```

```python
#create csv
songs.to_csv("csvs/songs.csv")
```

# Rank

```python
rank = main[["song_name", "spotify_rank", "apple_rank", "spotify_list_date", "apple_date_retrieved"]]
rank.head()
```

```python
rank = pd.merge(rank, songs, on="song_name")
rank.head()
```

```python
rank = rank.drop(["song_name","artist_ID","album_ID", "song_length"], axis=1)
rank.head()
```

```python
rank.info()
```

```python
#create csv
rank.to_csv("csvs/ranks.csv")
```
