```python
import pandas as pd
import numpy as np
```

```python
#read in spotify data
path = "Spotify_data.csv"

spotify = pd.read_csv(path)
spotify.head()
```

```python
#read in apple data
path = "apple_data.csv"

apple = pd.read_csv(path)
apple.head()
```

```python
#join together csv files
main = pd.merge(spotify, apple, on="song_name", how="outer")
main.head()
```

```python
#fill in artist column from two original columns
#https://stackoverflow.com/questions/55117802/how-can-i-check-if-a-value-is-nan-or-a-list
for i, row in main.iterrows():
    value = row.artist_y
    if pd.isnull(value):
        main["artist"]= main.artist_x
main.head()
```

```python
main.info()
```

```python
#review rows that have no artist
mask = main.artist.isnull()
main.loc[mask]
```

# artist table

```python
spotify_artist = list(spotify.artist)
```

```python
apple_artist = list(apple.artist)
```

```python
artists_org = spotify_artist + apple_artist
artists_org
```

```python
#create csv
artists.to_csv("final_csvs/artists.csv")
```

# Songs

```python
spotify_songs = list(spotify.song_name)
apple_songs = list(apple.song_name)
songs_org = spotify_songs + apple_songs
songs_org
```

```python
len(songs_org)
```

```python
len(np.unique(songs_org))
```

```python
songs_list = list(np.unique(songs_org))
```

```python
songs = pd.DataFrame()
songs["song_name"] = songs_list
songs.head()
```

```python
songs["song_ID"] = songs.index +1
songs.head()
```

```python
len(songs)
```

```python
#merge ID onto main 
main3 = pd.merge(main2, songs, on="song_name", how="left")
main3.head()
```

```python
main3.info()
```

```python
#retain only song specific information from main joined table
main_songs = main3[["song_name", "artist_ID", "album_ID", "song_length"]]
main_songs.head()
```

```python
#merge into songs df
songs = pd.merge(songs, main_songs, on="song_name")
songs.head()
```

```python
#create csv
songs.to_csv("final_csvs/songs.csv")
```

# Rank

```python
spotify_rank= pd.DataFrame()
spotify_rank["song_ID"] = main3["song_ID"]
spotify_rank["provider"] = main3["provider_x"]
spotify_rank["rank"] = main3["spotify_rank"]
spotify_rank["date_retrieved"] = main3["spotify_list_date"]
spotify_rank.head()
```

```python
mask = spotify_rank.provider =="spotify"
spotify_rank = spotify_rank.loc[mask]
spotify_rank.head()
```

```python
spotify_rank = pd.merge(spotify_rank, provider, on="provider", how="left")
spotify_rank
```

```python
apple_rank= pd.DataFrame()
apple_rank["song_ID"] = main3["song_ID"]
apple_rank["provider"] = main3["provider_y"]
apple_rank["rank"] = main3["apple_rank"]
apple_rank["date_retrieved"] = main3["apple_date_retrieved"]
apple_rank.head()
```

```python
mask = apple_rank.provider =="apple"
apple_rank = apple_rank.loc[mask]
apple_rank.head()
```

```python
apple_rank = pd.merge(apple_rank, provider, on="provider", how="left")
apple_rank
```

```python
ranks = pd.concat([spotify_rank, apple_rank], ignore_index=True)
ranks
```

```python
ranks = ranks.drop(columns=["provider"])
ranks.head()
```

```python
ranks
```

```python
#create csv
ranks.to_csv("final_csvs/ranks.csv")
```
